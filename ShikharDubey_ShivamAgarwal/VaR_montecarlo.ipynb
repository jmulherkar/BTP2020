```
import fix_yahoo_finance as yf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tabulate import tabulate 
import datetime as dt
# Import the price data of General Motors

tickers = ['AAPL','FB', 'C', 'DIS']
tickers = ['AAPL','FB', 'C', 'DIS']
df = yf.download(tickers, start='2012-01-01',end=dt.date.today())['Close']
df=df.dropna()
returns = df.pct_change()
```

```
weights = np.array([0.25,0.3,0.15,0.3])
```

```
mu = returns.mean(axis=0)
```

```
sigma = returns.std(axis=0)
```

```
mu = np.array(mu)
sigma = np.array(sigma)
```

```
ini_investment = 100000
```

```
investment = ini_investment*weights
```

```
investment
```

```
sigma
```

```
simulation = {}
for i in range(4):
  for sim in range(1,5):
      simulation[f"{i}_Simulation_"+str(sim)] = [investment[i]]
      for days in range(10):
          next_day = simulation[f"{i}_Simulation_"+str(sim)][-1]*np.exp((mu[i]-(sigma[i]**2/2))+sigma[i]*np.random.normal())
          simulation[f"{i}_Simulation_"+str(sim)].append(next_day)
```

```
simulation
```

```
AAPL_sim1 = np.array(simulation['0_Simulation_1'])
AAPL_sim2 = np.array(simulation['0_Simulation_2'])
AAPL_sim3 = np.array(simulation['0_Simulation_3'])
AAPL_sim4 = np.array(simulation['0_Simulation_4'])
FB_sim1 = np.array(simulation['1_Simulation_1'])
FB_sim2 = np.array(simulation['1_Simulation_2'])
FB_sim3 = np.array(simulation['1_Simulation_3'])
FB_sim4 = np.array(simulation['1_Simulation_4'])
C_sim1 = np.array(simulation['2_Simulation_1'])
C_sim2 = np.array(simulation['2_Simulation_2'])
C_sim3 = np.array(simulation['2_Simulation_3'])
C_sim4 = np.array(simulation['2_Simulation_4'])
DIS_sim1 = np.array(simulation['3_Simulation_1'])
DIS_sim2 = np.array(simulation['3_Simulation_2'])
DIS_sim3 = np.array(simulation['3_Simulation_3'])
DIS_sim4 = np.array(simulation['3_Simulation_4'])
```

```
sim1 = AAPL_sim1 + FB_sim1 + C_sim1 + DIS_sim1
sim2 = AAPL_sim2 + FB_sim2 + C_sim2 + DIS_sim2
sim3 = AAPL_sim3 + FB_sim3 + C_sim3 + DIS_sim3
sim4 = AAPL_sim4 + FB_sim4 + C_sim4 + DIS_sim4
```

```
# Plotting the actual Adjusted Close price of General Motors
day = np.arange(0,11,step=1)
plt.plot(day,sim1,day,sim2,day,sim3,day,sim4)
plt.xlabel('Day')
plt.ylabel('Portfolio_Value')
```

```
a = {}
for i in range(4):
  inv_val = investment[i]
  a["investment_"+str(i)] = [inv_val]
  for num in range(2000):
    val = inv_val*np.exp((mu[i]-(sigma[i]**2/2))+sigma[i]*np.random.normal())
    a["investment_"+str(i)].append(val)
```

```
a
```

```
inv_sum =  np.array(a["investment_0"])+np.array(a["investment_1"]) + np.array(a["investment_2"])+  np.array(a["investment_3"])
print(len(inv_sum))
```

```
new_inv_sum = inv_sum - np.sum(investment)
```

```
plt.hist(new_inv_sum,bins=40)
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```
val1 = (np.quantile(new_inv_sum,0.1))*(-1)
val2 = (np.quantile(new_inv_sum,0.05))*(-1)
val3 = (np.quantile(new_inv_sum,0.01))*(-1)
print( tabulate([['90%', val1], ['95%' , val2],['99%', val3]], headers=['Confidence Level value', 'Value at risk']))
```

```
n_day = [val3]
for i in range(10):
  n_day.append(n_day[0]*np.sqrt(i+2))
```

```
n_day
```

```
x_axis = ['day1','day2','day3','day4','day5','day6','day7','day8','day9','day10','day11']
plt.plot(x_axis,n_day)
plt.ylabel("N-day VaR")
plt.show()
```
