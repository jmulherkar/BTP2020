```
from scipy.io import loadmat ,savemat
inv_hist = loadmat('historical_inv_sum.mat')
inv_hist = inv_hist['array']
inv_monte = loadmat('montecarlo_inv_sum.mat')
inv_monte = inv_monte['array']
nday_hist = loadmat('historical_nday_var.mat')
nday_hist = nday_hist['array']
nday_monte = loadmat('montecarlo_nday_var.mat')
nday_monte = nday_monte['array']
```

```
import matplotlib.pyplot as plt

x_axis = ['Day1','Day2','Day3','Day4','Day5','Day6','Day7','Day8','Day9','Day10','Day11']
plt.plot(x_axis,nday_monte.reshape((nday_monte.shape[1])),label='Montecarlo')
plt.plot(x_axis,nday_hist.reshape((nday_hist.shape[1])),label='Historical')
plt.ylabel("N-day VaR")
plt.legend()
plt.show()
```

```
legend = ['Montecarlo', 'Historical']
plt.hist([inv_monte.reshape((inv_monte.shape[1])), inv_hist.reshape((inv_hist.shape[1]))], color=['blue', 'orange'],bins=40)
plt.xlabel("Returns")
plt.ylabel("Frequency")
plt.legend(legend)
plt.show()
```
